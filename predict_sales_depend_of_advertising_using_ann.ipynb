```python
import pandas as pd
from tensorflow.keras.layers import Dense,Input,BatchNormalization
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
import seaborn as sns
```

```python
df=pd.read_csv("./sample_data/advertising.csv")
df.head()
```

```python
x=df[["TV","Radio","Newspaper"]]
y=df["Sales"]
```

```python
x_train,x_test,y_tain,_y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=Sequential()
model.add(Input(shape=(3,)))
model.add(Dense(units=8,activation="relu",use_bias=True))
model.add(Dense(units=4,activation="relu",use_bias=True))
model.add(Dense(1))
```

```python
model.compile(optimizer="adam",loss="mean_absolute_error",metrics=["mean_absolute_error"])
```

```python
model.fit(x_train,y_tain,epochs=1000,batch_size=5,validation_split=0.2)
```

```python
model.predict(x_test)
```

```python
import matplotlib.pyplot as plt

plt.plot(model.history['loss'], label='Train Loss')
plt.plot(model.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```
